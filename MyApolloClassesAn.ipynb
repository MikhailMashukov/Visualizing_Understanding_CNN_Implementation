```python
rowCount = 15
denseExcLayerNames = []
for name in controlObj.netWrapper.net.getAllLayers():
    if name.find('dense_exc_') == 0:
        denseExcLayerNames.append(name)
```

```python
plotNum =  1
plt.figure(figsize=(15, rowCount * 2.5), dpi=150)
minMaxVals = {}
stdDevVals = {}
layerNames = denseExcLayerNames + ['conv_1', 'conv_212', 'final_conv_21',
                                'start_31', 'conv_312', 'final_conv_31', 'conv_322', 'final_conv_32', 
                                'final_conv_42', 'final_conv_51', 
                                'dense_1']
try:
    for epochNum in controlObj.netWrapper.getSavedNetEpochs():
        if epochNum < 2 or (epochNum > 1 and epochNum % 2 != 0):
            continue
#         print(epochNum)

        for layerName in layerNames:
            imageNum = 2
            # img = controlObj.imageDataset.getImage(controlObj.curImageNum, 'cropped', 'train')
            # plt.imshow(img / 255.0)
    #         activations, drawMode, stdData = controlObj.getActivationsData(epochNum, imageNum, layerName)
        #     img = controlObj.onShowActivationsPressed()
            activations = controlObj.netWrapper.getImageActivations(layerName, imageNum, epochNum)
#             print('%s activations: %s, min %.4f, max %.4f (%s)' % \
#                         (layerName, str(activations.shape), activations.min(), activations.max(),
#                          str([int(v[0]) for v in np.where(activations == activations.max())])))

            imageNum = 4
            activations2 = controlObj.netWrapper.getImageActivations(layerName, imageNum, epochNum)
            if not layerName in minMaxVals:
                minMaxVals[layerName] = []
                stdDevVals[layerName] = []
            minMaxVals[layerName].append([activations.min(), activations.max(),
                         activations2.min(), activations2.max()])
            stdDevVals[layerName].append([np.std(activations), np.std(activations2)])
except Exception as ex:
    print('Error: ', str(ex))        
    
for layerName in layerNames:        
    ax = plt.subplot(rowCount, 3, plotNum)
    ax.plot(minMaxVals[layerName])
    ax.set_ylim([y / 1.0 for y in ax.get_ylim()])
    ax = ax.twinx()
    ax.plot(stdDevVals[layerName], linestyle='-.')
    
    lims = ax.get_ylim()
    mean = np.mean(lims)
    diff = lims[1] - lims[0]
#     ax.set_ylim([mean - diff * 0.75, mean + diff * 2])
    plt.text(0.7, 0.8, layerName, transform = ax.transAxes)
    plotNum += 1
plt.show()
controlObj.netWrapper.saveCacheState()
```
